```python
covar=lambda x,y,v='ens': ((x-x.mean(dim=v))*(y-y.mean(dim=v))).mean(dim=v) 
corr=lambda x,y,v='ens': covare(x,y)/x.std(dim=v)/y.std(dim=v)
slope=lambda x,y,v='ens': covar(x,y)/x.std(dim=v)**2

# global avg. for lat-only dataarray
def gavglat(x):
    ll=np.cos(np.deg2rad(x.lat)).where(x > -999e9)
    x1=(x*ll).sum(dim='lat')/ll.sum(dim='lat')

    return x1
```

```python
def plotSinLat(datax,ax,**kwargs):
    """plotSinLat(datax): line plot vs. sin(latitude)
    datax is an xarray"""
    
    sca(ax)
    
    xlat=[-80,-50,-30,-15,0,15,30,50,80]
    xloc=[np.sin(np.deg2rad(ii)) for ii in xlat]
    xlab=[str(ii) for ii in xlat]

    plot(np.sin(np.deg2rad(datax.lat)),datax,**kwargs)
    xticks(xloc,xlab)
```

```python
def plot2d(d1,axx):
    d11=d1.copy()
    d11[0,0]=9e9
    d11[0,1]=-9e9
    
    ran=np.percentile(np.abs(d11),95)
#     ran=0.6
    n1=3
    n2=ran/(n1*2+1)
    lev1=(np.arange(n1+1))*(n2*2)+n2
    lev1=np.concatenate([-lev1[::-1],lev1])
    
    d11.plot.contourf(ax=axx,transform=ccrs.PlateCarree(),cmap=plt.cm.RdBu_r,levels=lev1)
    axx.coastlines()
    axx.set_aspect('auto', adjustable=None)
```

```python
## make plots pretty
def colorplot(ax):
    
    almost_black = '#262626' # tweek from Thorsten
    
    for spx in ['top','right']:
            ax.spines[spx].set_visible(False)

    # Only show ticks on the left and bottom spines
    ax.yaxis.set_ticks_position('left')
    ax.xaxis.set_ticks_position('bottom')
        
    for spx in ['bottom','left']:
        ax.spines[spx].set_linewidth(0.5)
        ax.spines[spx].set_color(almost_black)
    
    for ticks in ax.xaxis.get_ticklines() + ax.yaxis.get_ticklines():
        ticks.set_color(almost_black)

def colorplotRight(ax):
    
    almost_black = '#262626' # tweek from Thorsten
    
    for spx in ['top','left']:
            ax.spines[spx].set_visible(False)

    # Only show ticks on the left and bottom spines
    ax.yaxis.set_ticks_position('right')
    ax.xaxis.set_ticks_position('bottom')
    
    # set axis positions
#     ax.yaxis.set_label_position(pspines[0]) 
#     ax.xaxis.set_label_position(pspines[1]) 
    
    for spx in ['bottom','right']:
        ax.spines[spx].set_linewidth(0.5)
        ax.spines[spx].set_color(almost_black)
    
    for ticks in ax.xaxis.get_ticklines() + ax.yaxis.get_ticklines():
        ticks.set_color(almost_black)
    
    ax.yaxis.set_label_position('right')
```

```python
## read in cloud feedback for all individual ensemble members
ens=[]
for ii in range(1,69):

    try:
        x1=xr.open_dataset('/sn1/dessler/MPI-1percent/fluxes{:04d}.nc'.format(ii))
    except FileNotFoundError:
        break

    x1.expand_dims('ens')
    x1['ens']=[ii]

    x1=x1.drop('planck,lapserate,planckRH,lapserateRH,watervapor,watervaporRH,albedo,total'.split(','))
    ens.append(annual(x1))
    print(ii,end=" ")

# combine individual ensemble members
fluxes_individual=xr.concat(ens,dim='ens');ens=[]
fluxes_individual['time']=range(150)
fluxes_individual=fluxes_individual.assign(cloud=fluxes_individual.cloudsw+fluxes_individual.cloudlw)
# fluxes_individual.to_netcdf('/sn1/dessler/MPI-1percent/fluxes_gavg_all.nc')
```

```python
nvarx='cloud'

fig=figure(figsize=(9,4.5))
n1=4 # determines number of levels (= 2n1+1)

period1=(30,110);period2=(0,10)

ax1=fig.add_subplot(1,1,1,projection=ccrs.Robinson())   

g2=a4xco2[nvarx].mean(dim='time')/t4.mean() # feedback from 4xCO2 run

# feedback from individual 1% ensemble members
d2=(fluxes_individual[nvarx].sel(time=slice(*period1)).mean(dim='time')-fluxes_individual[nvarx].sel(time=slice(*period2)).mean(dim='time'))/\
   (gavg(fluxes_individual.temperature).sel(time=slice(*period1)).mean(dim='time')-gavg(fluxes_individual.temperature).sel(time=slice(*period2)).mean(dim='time'))
d1=g2-d2 # difference
g1=d2.mean(dim='ens') # ensemble average 

x1=g2-g1 # ensemble average difference
y1=x1.sel(lon=0)
y1['lon']=360
x2=xr.concat([x1,y1],dim='lon') # x2 = field to be plotted

# calculate robustness
thresh=68*0.9 # how many members have the same sign
y1=d1.where(d1 > 0).count(dim='ens')
y1=y1.where(y1 > thresh,0)
y2=d1.where(d1 < 0).count(dim='ens')
y2=-y2.where(y2 > thresh,0)
y3=np.sign(y1)+np.sign(y2)

x2.name='$\Delta \lambda_{'+nvarx+'}$'

x2[0,0]=9e9;x2[0,1]=-9e9

ran=np.percentile(np.abs(x1),95)

n2=ran/(n1*2+1)
lev1=(np.arange(n1+1))*(n2*2)+n2
lev1=np.concatenate([-lev1[::-1],lev1])
# lev1=[-1.6,-1.2,-0.8,0.8,1.2,1.6]
lev1=[-1.5,-1.,-0.5,0.5,1.,1.5]
# lev1=[-2.4,-1.6,-0.6,0.6,1.6,2.4]

x2.plot.contourf(ax=ax1,transform=ccrs.PlateCarree(),cmap=plt.cm.RdBu_r,levels=lev1)

y3.plot.contour(ax=ax1,transform=ccrs.PlateCarree(),levels=[-0.9,0.9],colors='0.4',linestyles='solid')

ax1.coastlines()
ax1.set_aspect('auto', adjustable=None)

ax1.set_title('',fontsize=18);

savefig('../figures/fig11.pdf')
```
